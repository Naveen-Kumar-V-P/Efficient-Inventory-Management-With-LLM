```python
import pandas as pd
from transformers import BertForQuestionAnswering, BertTokenizer
import torch

# Function to load and preprocess data
def load_data(csv_file):
    data = pd.read_csv(csv_file, encoding='latin1')
    # Preprocess the data if needed (e.g., handle missing values, clean text)
    return data

# Function to preprocess question
def preprocess_question(question):
    # Preprocess the question if needed (e.g., remove punctuation, lowercase)
    return question.strip()

# Function to answer a question using BERT model
def answer_question(question, data, model, tokenizer):
    question = preprocess_question(question)
    for index, row in data.iterrows():
        if question.lower() in row['Questions'].lower():
            return row['Answers']
    return "Sorry, I couldn't find an answer to that question."

# Function to initialize BERT model and tokenizer
def initialize_bert_model(model_name):
    tokenizer = BertTokenizer.from_pretrained(model_name)
    model = BertForQuestionAnswering.from_pretrained(model_name)
    return model, tokenizer

# Main function to interact with the user
def main():
    # Load CSV file
    csv_file = "/content/drive/MyDrive/sales.csv"
    data = load_data(csv_file)

    # Load pre-trained BERT model and tokenizer
    model_name = "bert-large-uncased-whole-word-masking-finetuned-squad"
    model, tokenizer = initialize_bert_model(model_name)

    # Infinite loop to interact with the user
    while True:
        # Prompt user to input a question
        question = input("Enter your question (or type 'quit' to exit): ")

        # Check if the user wants to quit
        if question.lower() == "quit":
            print("Exiting...")
            break

        # Answer the question based on the dataset
        answer = answer_question(question, data, model, tokenizer)
        print("Answer:", answer)

# Entry point of the script
if __name__ == "__main__":
    main()
```
